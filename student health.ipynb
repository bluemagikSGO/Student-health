```python
import pandas as pd
```

```python
df = pd.read_csv('Impact_of_Mobile_Phone_on_Students_Health.csv')
```

```python
df.head(5)
```

```python
df.count()
```

```python
df['Mobile Operating System '].unique()
```

```python
## Sort the values in the table so it be grouped by the name column in ascending order
```

```python
df = df.sort_values(by='Names')
```

```python
#change the column name from 'Gender ' to 'Gender' and 'Mobile Operating System ' to 'Mobile Operating System'
```

```python
df = df.rename(columns={'Gender ': 'Gender', 'Mobile Operating System ':'Mobile Operating System'})
```

```python
###change the value of the gender from Male to M (to save space)
```

```python
replacement = {'Male':'M', 'Female':'F'}

df['Gender'] = df['Gender'].replace(replacement)
```

```python
df.head()
```

```python
#work on the [usage symptoms] column
```

```python
df['Usage symptoms'].unique()
```

```python
replacement = {'Sleep disturbance;Anxiety or Stress':'Sleep disturbance|Anxiety or Stress','All of these':'Headache|Sleep disturbance|Anxiety or Stress', 'Headache;Sleep disturbance;Anxiety or Stress;All of these':'Headache|Sleep disturbance|Anxiety or Stress'}
df['Usage symptoms'] = df['Usage symptoms'].replace(replacement)
```

```python
df['Usage symptoms'].unique()
```

```python
#work on the [Mobile phone activities] column
```

```python
df.head()
```

```python
df['Mobile phone activities'].unique()
```

```python
replacement = {'Social Media;Messaging':'Social Media|Messaging', 'All of these':'Social Media|Web-browsing|Messaging','Social Media;Web-browsing;Messaging;All of these':'Social Media|Web-browsing|Messaging','Social Media;All of these':'Social Media|Web-browsing|Messaging'}
df['Mobile phone activities'] = df['Mobile phone activities'].replace(replacement)
```

```python
#work on the [Daily usages] column
```

```python
df.head()
```

```python
df['Daily usages'].unique()
```

```python
replacement = {'> 6 hours':'6+ hours', '< 2 hours':'-2 hours'}
df['Daily usages']=df['Daily usages'].replace(replacement)
```

```python
#work on the [Health precautions] column
```

```python
df['Health precautions'].unique()
```

```python
df['Health precautions']=df['Health precautions'].replace({'None of Above':'No Precaution'})
```

```python
#work on the [Health rating] column
```

```python
df['Health rating'].unique()
```

```python
df['Health rating'] = df['Health rating'].replace({'Excellent;Good':'Excellent|Good','Good;Fair':'Good|Fair','Excellent;Good;Fair;Poor':'Excellent|Good|Fair|Poor'})
```

```python
df['Health rating'].unique()
```

```python
df.head()
```

```python
df[df.isna().any(axis=1)]
```

```python
df[df.isna().sum(axis=1) > 2]
```

```python
df = df[df.isna().sum(axis=1) <= 2]
```

```python
df.head()
```

```python
df.to_csv('transformed.csv', index=False)
```
